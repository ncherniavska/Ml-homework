**1. Списки - базові операції**

Створіть список температур за тиждень (у градусах Цельсія) з назвою
`temperatures` і значеннями [22, 25, 19, 23, 26, 20, 21, 22].


Виконайте (кожен пункт оцінюється в 0.5 бала):
- Виведіть перший елемент
- Виведіть останній елемент
- Додайте температуру 24
- Видаліть температуру 19
- Виведіть довжину списку
- Обчисліть суму всіх температур
- Знайдіть максимальну температуру

```python
# Створення списку
temperatures = [22, 25, 19, 23, 26, 20, 21, 22]
```

```python
# Виводимо перший елемент
print (temperatures[0])
```

```python
# Виводимо останній елемент
print (temperatures[-1])
```

```python
# Додаємо температуру 24 в кінець списку
temperatures.append(24)
print(temperatures)
```

```python
# Видаляємо елемент 19 зі списку
temperatures.remove(19)
print(temperatures)
```

```python
# Корисні методи
print(len(temperatures)) # Довжина списку після додавання й видалення елементу
print(max(temperatures)) # Максимальне значення списку
print(sum(temperatures)) # Сума всіх значень списку
```


**2. Словники - базові операції**

Створіть словник про себе, який містить ключі `name`, `age`, `city` і значення, які відповідають вашему імені, віку і місту (можна вставити будь-які значення, не обовʼязково свої).


Виконайте (кожен пункт оцінюється в 0.5 бала):
- Виведіть значення ключа "name"
- Змініть значення ключа "age" на 21
- Додайте новий ключ "hobby" зі значенням "programming"
- Виведіть всі ключі
- Виведіть всі значення


```python
# Створення списку

student = {
    "name": "Nadiia",
    "age": 36,
    "city": "Calgary"
}
```

```python
# Виводимо значення ключа "name"

student['name']
```

```python
# Змінюємо значення ключа "age" на 21

student['age'] = 21
student['age']
```

```python
# Додаємо новий ключ "hobby" зі значенням "programming"

student["hobby"] = "programming"
student
```

```python
# Виводимо всі ключі
print(student.keys())
```

```python
# Виводимо всі значення
print(student.values())
```
