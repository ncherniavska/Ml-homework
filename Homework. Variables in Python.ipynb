**2. Арифметичні операції**

Створіть дві змінні: `var1` зі значенням 15, `var2` зі значенням 4.

Обчисліть та виведіть (кожен пункт оцінюється в 0.5 бала):
- Суму
- Різницю
- Добуток
- Ділення
- Цілочисельне ділення
- Залишок від ділення
- Піднесення до степеня


```python
# Arithmetic operations
var1 = 15
var2 = 4

print(var1 + var2)  # addition
print(var1 - var2)  # substraction
print(var1 * var2)  # multiplication
print(var1 / var2)  # division
print(var1 // var2)  # integer devision
print(var1 ** var2)  # exponentiation
print(var1 % var2)  # modulo
```
